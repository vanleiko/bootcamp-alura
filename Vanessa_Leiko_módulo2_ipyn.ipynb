**Comparação da cobertura vacinal entre os estados do Amazonas e São Paulo**


Resolvi comparar a cobertura vacinal entre os anos de 2014 e 2018 nos estados do Amazonas e de São Paulo porque são dois estados que têm realidades e características bem distintas. Meu objetivo foi avaliar se esses aspectos poderiam influenciar no cumprimento da meta da cobertura vacinal.

A fórmula de cálculo da cobertura é o número de doses aplicadas da dose indicada (1ª, 2ª, 3ª dose ou dose única, conforme a vacina) dividida pela população-alvo, multiplicado por 100. 

Analisando a distribuição dos valores das taxas de cobertura vacinal, no estado do Amazonas esses valores se concentraram entre 70% e 90%, enquanto que em São Paulo, a concentração das taxas ficou entre 90% e 100%.

Dois destaques são:

a) a baixa taxa de vacinação da febre amarela no estado de São Paulo entre os anos de 2014 e 2017, seguido de um rápido aumento em 2018, podendo estar relacionado com o surto da doença que a região Sudeste viveu nesse período;

b) a queda na taxa de vacinação contra o sarampo no estado do Amazonas, podendo também ter relação com o surto dessa doença no estado em 2018.

**Criando uma função para formatar as tabelas**

```
def formata_tabela(tabela, estado):

  # renomeia as colunas
  df = tabela.rename(columns={0: "Indicador", 1: estado})

  # cria uma lista com os indicadores escolhidos
  lista_indicadores = list(df.Indicador)

  # cria um dicionário com os indicadores e remove o "-" da frente
  nome_indicador = {item: item[2:] for item in lista_indicadores}

  # substituindo os nomes dos indicadores
  df["Indicador"] = df["Indicador"].map(nome_indicador)

  # formatando o indicador "Clima" que perdeu o "Cl" no passo anterior 
  df["Indicador"].replace(to_replace={"Total": "Area", "Estimativa": "Estimativa populacional", "ima": "Clima"}, inplace=True)
  
  # organizando o index
  df.reset_index(drop=True, inplace=True)

  return df
```

```
# chamando a função para formatar a tabela do Amazonas
tabela_am = formata_tabela(selecao_am, "Amazonas")

# ajustando o valor do Clima (para ficar menor)
tabela_am["Amazonas"].replace({"Equatorial e tropical de monção Am, Af": "Equatorial e tropical de monção"}, inplace=True)
tabela_am
```

```
# chamando a função para formatar a tabela de São Paulo
tabela_sp = formata_tabela(selecao_sp, "Sao Paulo")

# renomeando o valor do Clima (para ficar menor)
tabela_sp["Sao Paulo"].replace({"Subtropical, tropical de altitude e tropical Cfa, Cfb, Cwa, Cwb, Aw": "Subtropical, trop. de altitude e tropical"}, inplace=True)
tabela_sp
```

**Função para formatar os dataframes**



```
# função que irá formatar o dataframe

def formata_df(df):
  
  # pegando os nomes das vacinas na coluna 'Imuno'
  nomes_vacinas = list(df["Imuno"]) 

  # removendo o número das vacinas e criando um dicionário
  vacinas = {nome: nome[4:] for nome in nomes_vacinas}  

  # renomeando os nomes das vacinas 
  df["Imuno"] = df["Imuno"].map(vacinas)

  # removendo a coluna Totalnomeando as colunas
  df.drop("Total", axis=1, inplace=True)

  return df
```

```
# chamando a função para formatar os dados do Amazonas
cobertura_am = formata_df(dados_cobertura_am)
cobertura_am.head()
```

```
# chamando a função para formatar os dados de São Paulo
cobertura_sp = formata_df(dados_cobertura_sp)
cobertura_sp.head()
```

**Melt nos dataframes**



Usei a função melt para colocar os nomes dos estados como valores do dataframe

```
def melt_df(df):

  # colocando os anos como valores do df
  df = df.melt(id_vars="Imuno", var_name="Ano", value_name="Cobertura")

  # colocar Imuno como índice
  df = df.set_index("Imuno")

  return df.sort_index()
  
```

```
# chamando a função melt para o Amazonas
cobertura_am_melt = melt_df(cobertura_am)
cobertura_am_melt.head(10)
```

```
# chamando a função melt para São Paulo
cobertura_sp_melt = melt_df(cobertura_sp)
cobertura_sp_melt.head(10)
```

```
# os decimais do dataframe estavam separados tanto por vírgula quanto por ponto, além de ter valores preenchidos com ',,,'
# tive que criar essa função para trocar as virgulas dos decimais por ponto, e trocar os ',,,' por zero

def formata_decimal(dado):

  dict_novos_valores = {}

  lista_valores = list(dado["Cobertura"])

  for valor in lista_valores:
    if valor == ',,,':
      novo_valor = valor.replace(valor,'0')
      dict_novos_valores[valor] = float(novo_valor)
    else:
      novo_valor = str(valor).replace(',','.')
      dict_novos_valores[valor] = float(novo_valor)
  
  dado["Cobertura"] = dado["Cobertura"].map(dict_novos_valores)
  dado["Ano"] = dado["Ano"].astype(int)

  return dado

```

```
# chamando a função para corrigir os decimais
cobertura_sp_melt = formata_decimal(cobertura_sp_melt)
cobertura_am_melt = formata_decimal(cobertura_am_melt)
```

###**Histograma - Distribuição da cobertura vacinal**

Optei por criar um histograma da cobertura vacinal para ver a frequência de seus valores.

Analisando os histogramas abaixo, é possível verificar que o estado de São Paulo possui a maioria das suas taxas de cobertura vacinal entre 90% e 100%, enquanto que no Amazonas, a maior parte das taxas estão entre 70% e 90%. Essa menor taxa no Amazonas pode estar relacionada a maior dificuldade de acesso do Estado até a população, visto que quase metade da população do estado vive na capital, enquanto que o restante vive distribuída nas cidades do interior e comunidades ribeirinhas e indígenas, sendo muitas dessas de difícil acesso e sem postos de saúde local, além da maioria dessas comunidades dependerem de gerador de energia, impossibilitando o estoque de vacinas.

```
# função para criar um histograma para ver a distribuição da cobertura vacinal nos estados

def cria_histograma(dados, estado, bins):
  
  plt.figure(figsize=(12,6))
  
  grafico = sns.histplot(data=dados, x="Cobertura", kde=True, bins=bins, zorder=1.1, )
  
  grafico.yaxis.grid(color="gray", linestyle="dashed")

  plt.xlabel("Taxa de cobertura vacinal (%)", fontsize=16, fontweight="bold")
  plt.ylabel("Contagem", fontsize=16, fontweight="bold")

  plt.xticks(fontsize=14)
  plt.xticks(range(0,120,10))
  plt.yticks(fontsize=14)
  plt.yticks(range(0,55,5))

  plt.title(f"Distribuição dos valores da cobertura vacinal\nno estado - {estado} - entre 2014 e 2018", fontsize=16, fontweight="bold")    
```

####<font color="darkblue">**Figura 1. Ditribuição da taxa de imunização no Amazonas entre 2014 e 2018**

```
# histograma da cobertura vacinal no Amazonas
cria_histograma(cobertura_am_melt, "Amazonas", 12)
```

####<font color="darkblue">**Figura 2. Ditribuição da taxa de imunização em São Paulo entre 2014 e 2018**

```
# histograma da cobertura vacinal em São Paulo
cria_histograma(cobertura_sp_melt, "São Paulo", 11)
```

###**Gráfico de barras - Cobertura vacinal entre 2014 e 2018**

Fiquei na dúvida se plotar todas as vacinas em um único gráfico seria uma boa prática de visualização dos dados (visto que ficou meio poluído). Mas eu optei por fazer isso para ter uma visão macro de como foram as taxas de cobertura vacinal de cada vacina ao longo desses 5 anos.

```
# função para plotar o gráfico de barras
def cria_grafico(dados, estado):
  plt.figure(figsize=(14,6))
  
  grafico = sns.barplot(data=dados, x=dados.index, y="Cobertura", hue="Ano", zorder=1.1, palette="Set1")
  
  grafico.set_axisbelow(True)
  grafico.yaxis.grid(color='gray', linestyle='dashed')

  plt.ylabel("Cobertura vacinal (%)", fontsize=16, fontweight="bold")
  plt.xlabel("Vacina", fontsize=18, fontweight="bold")

  plt.yticks(np.arange(0, 110, step=10), fontsize=14)
  plt.xticks(rotation=40, fontsize=14, horizontalalignment="right")

  plt.title(f"Taxa de imunização no estado - {estado} - entre 2014 e 2018", fontsize=18, fontweight="bold")
  plt.legend(title="Ano", bbox_to_anchor=(1.13,1), fontsize="x-large", title_fontsize=14);
```

####<font color="darkblue">**Figura 3.Taxa de imunização no estado de São Paulo entre 2014 e 2018**

```
cria_grafico(cobertura_sp_melt, "São Paulo")
```

Em relação a **São Paulo**, o que chamou a atenção foi a baixa cobertura vacinal contra a **febre amarela** entre 2014 e 2017, seguida de um aumento na taxa de cobertura em 2018, fato que pode estar relacionado com o [**surto**]("https://www.bio.fiocruz.br/index.php/br/noticias/1399-o-que-sabemos-sobre-o-surto-de-febre-amarela-no-brasil") da doença que o país viveu em 2016 e 2017. Isso indica que baixas taxas de cobertura vacinal podem proporcionar surtos de doenças nos anos seguintes. O acompanhamento das taxas de vacinação é uma importante forma de se monitorar, prever e prevenir possíveis surtos que são evitáveis por meio da vacinação.
