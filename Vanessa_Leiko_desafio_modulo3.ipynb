```
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2
import plotly.graph_objects as go
```

**Importação dos dados**

```
dados.isna().value_counts()
```

**Funções utilizadas nas análises**

<font color="blue">*Função para gerar crosstab de frequência:*

```
def gera_crosstab_frequencia(dados, variavel1, variavel2, titulo1, titulo2):

  df = pd.crosstab(dados[variavel1], dados[variavel2])

  df.rename(columns={1:titulo1, 2:titulo2}, inplace=True)   

  return df
```

<font color="blue">*Função para gerar crosstab de porcentagem, normalizada pelo índice:*

```
def gera_crosstab_porcentagem(dados, variavel1, variavel2, titulo1, titulo2):

  df = ((pd.crosstab(dados[variavel1], dados[variavel2], normalize="index")) * 100).round(2)

  df.rename(columns={1:titulo1, 2:titulo2}, inplace=True)   

  return df
```

<font color="blue">*Função para gerar gráfico:*

```
def gera_grafico(nome1, nome2, dados, variavel1, variavel2, titulo):

  fig = go.Figure(data=[
                        go.Bar(name=nome1, x=dados.index, y=dados[variavel1].values),
                        go.Bar(name=nome2, x=dados.index, y=dados[variavel2].values)
                        ])

  fig.update_layout(title=titulo, title_font_size=28, width=1200, height=500)

  fig.update_yaxes(title_text="Porcentagem (%)", range=[0,100], title_font_size=18)    

  fig.show()
```

<font color="blue">*Função para calcular o qui-quadrado:*

```
def calcula_qui_quadrado(df, variavel1, variavel2):

  qui_quadrado, p_valor, graus_de_liberdade, valores_esperados = stats.chi2_contingency(df)

  alfa = 0.05

  valor_critico = chi2.ppf(q=1-alfa, df=graus_de_liberdade)

 
  if qui_quadrado >= valor_critico:
    print(f'Há relação entre as variáveis "{variavel1}" e "{variavel2}" --> Rejeitamos a hipótese nula')
    print()
    print(f"p-value = {p_valor:.4f}")
    print(f"Alfa = {alfa}")
    print()
    print(f"Graus de liberdade = {graus_de_liberdade}")
    print(f"Valor de P da tabela = {valor_critico}")
    print(f"Valor qui-quadrado = {qui_quadrado}")

    

  else:
    print(f'Não há relação entre as variáveis "{variavel1}" e "{variavel2}" --> Aceitamos a hipótese nula')
    print()
    print(f"p-value = {p_valor:.4f}")
    print(f"Alfa = {alfa}")
    print()
    print(f"Graus de liberdade = {graus_de_liberdade}")
    print(f"Valor de P da tabela = {valor_critico}")
    print(f"Valor qui-quadrado = {qui_quadrado}")

```

**Teste Qui-quadrado:**

**H<sub>0</sub>: Sofrer bullying é igual para meninos e meninas.**

**Resultado: Rejeitamos a hipótese nula.**

```
# optei por realizar o teste apenas com quem respondeu "SIM" ou "NÃO"
df_sofreu_bullying_freq.iloc[[0, 1]]
```

```
calcula_qui_quadrado(df_sofreu_bullying_freq.iloc[[0, 1]], "Gênero", "Sofrer bullying")
```
